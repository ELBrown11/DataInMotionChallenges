# **Challenge 3: Alcohol Consumption**

```python
#import libraries
import pandas as pd
```

```python
#data url
url = 'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/drinks.csv'
```

```python
#read in data
data = pd.read_csv(url)
data.head()#preview data
```

# **Questions**

1. Which continent drinks more beer on average?

```python
avg_drinks = data.groupby(by='country')['total_litres_of_pure_alcohol'].mean().sort_values(ascending=False)
avg_drinks
```

2. For each continent print the statistics for wine consumption.

```python
wine_continent = data[['continent','wine_servings']]
wine_continent
```

```python
for i in range(len(wine_continent)):
  print(wine_continent.iloc[[i]])
```


3. Print the mean alcohol consumption per continent for every column.

```python
avg_drinks_cont = data.groupby(by='continent')['total_litres_of_pure_alcohol'].mean()
avg_drinks_cont
```

```python
for i in range(len(avg_drinks_cont)):
  print(avg_drinks_cont.iloc[[i]])
```

4. Print the median alcohol consumption per continent for every column.

```python
med_drinks_cont = data.groupby(by='continent')['total_litres_of_pure_alcohol'].median()
med_drinks_cont
```

```python
for i in range(len(med_drinks_cont)):
  print(med_drinks_cont.iloc[[i]])
```
